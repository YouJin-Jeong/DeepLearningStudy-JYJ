```python
AAA = np.array(f["Anaconda_t"][:])
print(AAA)
```

```python
BBB = 1000 * AAA
```

```python
CCC = BBB.transpose()
print(np.shape(CCC))
```

```python
DDD = np.zeros((np.size(AAA,1), np.size(CCC,1), np.size(AAA,0)))
for i in range(0, 3):
    DDD[:,:,i] = AAA[i,:,:]@CCC[:,:,i]
```

```python
N_DDD = DDD/(np.amax(DDD))
```

```python
plt.imshow(N_DDD)
plt.show()
```

```python
hf = h5py.File('DDD.h5', 'w')
hf.create_dataset('DDD', data = DDD)
```
